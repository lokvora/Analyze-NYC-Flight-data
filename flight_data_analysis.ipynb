```python
import pandas as pd
import numpy as np
```

```python
flight=pd.read_csv("flight_data.csv")
flight["dep_time_new"]= pd.to_numeric(flight['dep_time']).fillna(0).astype(np.int64).apply(lambda x: '{0:0>4}'.format(x))
flight["dep_time_new"]= pd.to_timedelta(flight["dep_time_new"].str[:2]+":"+flight["dep_time_new"].str[-2:]+":00")

flight["arr_time_new"]= pd.to_numeric(flight['arr_time']).fillna(0).astype(np.int64).apply(lambda x: '{0:0>4}'.format(x))
flight["arr_time_new"]=pd.to_timedelta(flight["arr_time_new"].str[:2]+":"+flight["arr_time_new"].str[-2:]+":00")

flight["sched_dep_time_new"]= pd.to_numeric(flight['sched_dep_time']).fillna(0).astype(np.int64).apply(lambda x: '{0:0>4}'.format(x))
flight["sched_dep_time_new"]=pd.to_timedelta(flight["sched_dep_time_new"].str[:2]+":"+flight["sched_dep_time_new"].str[-2:]+":00")

flight["sched_arr_time_new"]= pd.to_numeric(flight['sched_arr_time']).fillna(0).astype(np.int64).apply(lambda x: '{0:0>4}'.format(x))
flight["sched_arr_time_new"]=pd.to_timedelta(flight["sched_arr_time_new"].str[:2]+":"+flight["sched_arr_time_new"].str[-2:]+":00")

flight
```

# Best airport in terms of time departure %

```python
flight_count=pd.DataFrame(flight.groupby("origin")[["year"]].count())
delay_flight=pd.DataFrame(flight[flight.dep_delay>0].groupby("origin")[["dep_delay"]].count())
ontime_flight=pd.DataFrame(flight[flight.dep_delay<=0].groupby("origin")[["dep_delay"]].count())
cancelled_flight=pd.DataFrame(flight[flight["dep_delay"].isnull()].groupby("origin")[["sched_dep_time"]].count())
```

```python
flight_temp=pd.merge(delay_flight, ontime_flight, left_index=True, right_index=True, suffixes=('_delay', '_onTime'))
flight_temp2=pd.merge(flight_temp, cancelled_flight, left_index=True, right_index=True)
flight_dep=pd.merge(flight_temp2, flight_count, left_index=True, right_index=True)
flight_dep.rename(columns={"dep_delay_delay":"delayed","dep_delay_onTime":"ontime","sched_dep_time":"cancelled","year":"total"}, inplace=True)
flight_dep["delay%"]=flight_dep.delayed/flight_dep.total*100
flight_dep["ontime%"]=flight_dep.ontime/flight_dep.total*100
flight_dep["cancelled%"]=flight_dep.cancelled/flight_dep.total*100
```

```python
import matplotlib.pyplot as plt

print(flight_dep[["delay%","ontime%","cancelled%"]])

flight_dep[["delay%","ontime%","cancelled%"]].plot(kind ="bar", title = "Departure")
plt.ylabel("%")
plt.show()

print("LGA airport is best in departure time %")
```

# On time arrival % analysis

```python
mask=flight.arr_time.isna() & flight.dep_time.notna()
```

```python
flight.loc[mask,"arr_time_new"]=flight.dep_time_new+flight.sched_arr_time_new-flight.sched_dep_time_new
```

```python
flight_arr=flight.dropna(subset=["arr_time", "dep_time"])
flight_arr_count=pd.DataFrame(flight_arr.groupby("dest")[["year"]].count())

flight_LA=pd.DataFrame(flight_arr[flight_arr.arr_delay>0].groupby("dest")[["arr_delay"]].count())
flight_OT=pd.DataFrame(flight_arr[flight_arr.arr_delay<=0].groupby("dest")[["arr_delay"]].count())
flight_arrival=pd.merge(flight_LA, flight_OT, left_index=True, right_index=True)

#flight_arrival=pd.merge(flight_temp, flight_arr_count, left_index=True, right_index=True)

#flight_arrival
flight_arrival.rename(columns={"arr_delay_x":"delayed","arr_delay_y":"ontime"}, inplace=True)

flight_arrival["LA%"]=flight_arrival.delayed/(flight_arrival.delayed+flight_arrival.ontime)*100
flight_arrival["OT%"]=flight_arrival.ontime/(flight_arrival.delayed+flight_arrival.ontime)*100

flight_arrival.sort_values(by=["OT%"], ascending=False)
#print(len(flight_arr)) - 328063

print("Destination MVY is best in terms of arrival %")
```
